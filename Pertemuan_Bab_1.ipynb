---

## **2. Konsep Utama dalam ER Diagram**

Beberapa komponen dasar dalam ERD:

* **Entitas**: objek nyata yang digambarkan sebagai persegi panjang (contoh: Karyawan, Obat).
* **Atribut**: karakteristik dari entitas (contoh: Nama, Alamat).
* **Primary Key**: atribut unik yang mengidentifikasi entitas (contoh: NIM, KodeObat).
* **Relasi**: hubungan antar entitas (contoh: meminjam, memiliki).
* **Kardinalitas**: menunjukkan jumlah keterlibatan entitas dalam relasi, seperti 1:1, 1:N, atau M:N.

---

## **3. Aturan Konversi ER ke Skema Relasi**

Bab ini memberikan aturan rinci (beserta contoh) untuk mengubah berbagai bentuk ER menjadi tabel relasional:

### **a. Entitas Kuat (Strong Entity)**

* Menjadi **1 tabel**.
* Semua atribut sederhana menjadi kolom.
* Atribut kunci = primary key.

### **b. Atribut Komposit**

* Dipecah menjadi atribut-atribut penyusunnya.

### **c. Atribut Multivalue**

* Dibuatkan **tabel baru** untuk atribut tersebut.
* Primary key tabel kuat menjadi **foreign key**.

### **d. Atribut Turunan (Derived Attribute)**

* Boleh dijadikan kolom bila diperlukan (contoh: umur).

### **e. Entitas Lemah**

* Menjadi tabel sendiri.
* Primary key-nya terdiri dari **PK entitas kuat + atribut pembeda**.
* PK entitas kuat menjadi **foreign key**.

### **f. Relasi 1:1**

* Dibuat 2 tabel.
* Salah satu tabel menyimpan **foreign key** dari tabel pasangannya.

### **g. Relasi 1:N**

* Atribut kunci entitas "1" menjadi **foreign key** di entitas "N".

### **h. Relasi M:N**

* Dibuat **tabel baru** sebagai hasil relasi.
* Primary key tabel baru dapat berupa gabungan FK atau ID baru.

### **i. Relasi Unary**

* Bergantung kardinalitas, bisa menjadi FK pada tabel sendiri atau tabel relasi baru.

### **j. Relasi Ternary**

* Dibuatkan **tabel relasi** yang berisi foreign key dari semua entitas terlibat.

### **k. Generalisasi & Spesialisasi (Inheritance)**

Ada 2 metode:

1. **Supertype-subtype terpisah** (superclass = tabel sendiri, subclass = tabel masing-masing).
2. **Setiap subclass menjadi tabel lengkap** yang membawa atribut superclass.

### **l. Agregasi**

* Relasi antara relasi.
* Dibangun tabel relasi yang membawa semua foreign key dari elemen yang terlibat.
