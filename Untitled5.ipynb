# **Rangkuman Bab 1 – Konversi ER ke Skema Relasi**

Bab 1 membahas bagaimana **mengubah Entity Relationship (ER) Diagram** menjadi **skema relasi** dan kemudian menjadi **tabel fisik** dalam basis data. Materi ini penting sebagai dasar sebelum membangun database di level fisik.

---

## **1. Tujuan Pembelajaran**

Praktikan memahami:

1. Cara mengonversi **ER Diagram → Skema Relasi → Tabel Database**.
2. Mempermudah proses transformasi desain ke basis data fisik (MySQL).

---

## **2. Konsep Dasar dalam ERD**

* **Entitas** → objek nyata (misal: Karyawan, Pasien, Obat)
* **Atribut** → karakteristik entitas (nama, alamat, umur)
* **Primary Key (PK)** → atribut unik pembeda setiap entitas
* **Relasi** → hubungan antar entitas
* **Kardinalitas** → bentuk hubungan (1–1, 1–N, N–N)

---

## **3. Aturan Konversi ERD ke Skema Relasi**

### **A. Entitas**

1. **Entitas kuat → 1 tabel**

   * Semua atribut menjadi kolom
   * PK tetap menjadi PK tabel

2. **Atribut komposit**

   * Dipecah menjadi atribut-atribut sederhana

3. **Multivalue atribut**

   * Dibuatkan tabel baru
   * PK tabel berisi PK entitas + atribut multivalue

4. **Entitas lemah**

   * Menjadi tabel baru
   * Wajib menyertakan FK dari entitas induk

---

### **B. Relasi**

1. **Relasi 1–1**

   * Salah satu tabel menyimpan FK tabel lain
   * Bisa ada pilihan tabel mana yang memegang FK

2. **Relasi 1–N**

   * Tabel dengan kardinalitas N menyimpan FK dari tabel 1

3. **Relasi N–N**

   * Dibuat tabel relasi baru
   * Berisi FK dari kedua entitas
   * Bisa diberi PK baru (surrogate key)

4. **Relasi dengan atribut**

   * Atribut pada relasi menghasilkan tabel baru

5. **Relasi unary (self relationship)**

   * Pada 1–N: tabel menyimpan FK yang merujuk dirinya sendiri
   * Pada N–N: membuat tabel relasi baru

6. **Relasi ternary (tiga entitas)**

   * Membuat tabel relasi baru yang memuat FK dari 3 entitas

---

### **C. Generalisasi dan Spesialisasi (ISA)**

Ada dua metode:

1. **Metode 1**

   * Superclass → tabel
   * Subclass → tabel terpisah, PK subclass = PK superclass

2. **Metode 2**

   * Subclass → tabel lengkap (mengandung atribut superclass + subclass)
