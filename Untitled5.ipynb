# **Rangkuman BAB 2 – Pengantar Basis Data (DDL)**

### **1. Pengertian DBMS MySQL**

* **MySQL** adalah salah satu *Database Management System (DBMS)* berbasis SQL yang bersifat **open source**, cepat, stabil, dan banyak digunakan.
* Alasan menggunakan MySQL:

  * Performa cepat dan reliabel
  * Banyak digunakan dalam proyek besar
  * Gratis (open source)
  * Multi-platform
  * Mendukung multi-thread, enkripsi, dan autentikasi user

---

### **2. Instalasi MySQL**

* Umumnya MySQL diinstal melalui paket **XAMPP**, yang menyertakan MySQL/MariaDB dan phpMyAdmin.
* XAMPP versi terbaru menggunakan **MariaDB** tetapi perintah SQL tetap sama.

---

### **3. Aplikasi Server MySQL**

* Setelah instalasi, server MySQL dapat dijalankan melalui:

  * XAMPP Control Panel
  * File mysqld.exe (jika manual)
* User default: **root** (tanpa password).

---

### **4. Mengakses MySQL**

Akses menggunakan **Command Prompt**:

1. Masuk ke direktori MySQL:

   ```
   cd c:\xampp\mysql\bin
   ```

2. Login:

   ```
   mysql -u root -p
   ```

   (Password kosong jika belum diset)

3. Keluar:

   ```
   \q
   ```

Lokasi penyimpanan database:

* Windows: `c:\xampp\mysql\data`
* Linux: `/opt/lampp/var/mysql`

---

### **5. Client MySQL**

* Aplikasi **mysql.exe** adalah client untuk mengirim perintah SQL ke server.

Aturan penting:

* Semua perintah SQL harus diakhiri **titik koma (;)**.

---

### **6. Tipe Data MySQL**

Beberapa tipe data penting:

| Tipe Data | Fungsi           | Format                     |
| --------- | ---------------- | -------------------------- |
| INT       | Bilangan bulat   | -2147483648 s/d 2147483648 |
| FLOAT     | Bilangan desimal | -                          |
| DATE      | Tanggal          | YYYY-MM-DD                 |
| DATETIME  | Tanggal & waktu  | YYYY-MM-DD HH:MM:SS        |
| CHAR      | String tetap     | 1–255 karakter             |
| VARCHAR   | String variabel  | 1–255 karakter             |

---

### **7. Database Relasional**

* Database terdiri dari **tabel-tabel**.
* Untuk membuat tabel, harus membuat database terlebih dahulu.
* DDL (Data Definition Language) digunakan untuk membuat struktur database.

---

### **8. Perintah Dasar DDL**

#### **a. Membuat Database**

```
create database nama_database;
```

#### **b. Melihat Semua Database**

```
show databases;
```

#### **c. Mengakses Database**

```
use nama_database;
```

#### **d. Menghapus Database**

```
drop database nama_database;
```
